## 1. The Structure of Data

[note: The following is helpful for me to think about. If that's not true of you, you can skip this and get to the good stuff.]

The data that we collect, use math to manipulate, represent in spreadsheets and graphics, etc. is fundamentally no different from the information that gets processed by biological perceptual systems. There are, of course, machine perceptual systems whose input stimuli are energy distributions (like in biological systems) but the similarity with biological systems extends even to those machines whose perceptual 'stimuli' are spreadsheets. And this is because representation, whether biological or machine, has a _predicational_ structure. That is, the basic unit of representation is the application of a predicate __P__ to an object _a_: __P__(_a_). [note: for here and for all non-code in this notebook, I am using 'object' other terms in their _philosophical_ senses, not in whatever proprietary senses they may be used in, e.g., the pandas documentation.] If you are a realist, you'll think that this is to be expected: events in the world (understood in a general sense as actual states of affairs) are just property-instantiations. Thus, we describe events by identifying which properties are instantiated in which object(s) (if any). If we are accurately to represent the world, our representations ought to have the structure of events, and so must too be predicational in structure.

```python
df = pd.read_csv('abs-licenses-casnd-cleanerxy.csv', index_col='OGC_FID')
df[:4]
```

It may not be immediately obvious that this applies to spreadsheets, but consider that the above spreadsheet is just a collection of predications. The _objects_ are particular licenses, and the _predicates_ are several: status, issue date, expiration date, owner of establishment that has the license, location of establishment that has the license, etc. (strictly speaking, the predicates are not the column headings but particular values for those headings). Every single bit of information in that spreadsheet is an application of some predicate to some object, in this case a particular liquor license. In the present spreadsheet, particular licenses exist as points in an _n_-dimensional hyperspace, where _n_ is the number of columns and the dimensions are defined by the features described in the column labels (e.g., status, issue date, expiration date, etc.). Each object takes a value (numerical or not) along each dimension (ideally).   

It is customary, in a spreadsheet, to assign a row to each object and a column to each predicate, and this spreadsheet is no different. The rows in this spreadsheet are particular licenses, which we can identify uniquely and sequentially by their OGCFIDs. As such, the OGCFID makes a good _index column_ (if you do not specify an index column, it gives you a default index column, which is just the integers > -1, instead of the natural numbers, which are somewhat more 'natural' as indices). Think of the OGCFID like a name for each license: {'1','2',...,'n'}. The columns in the spreadsheet are predicates, some examples of which are listed above. No matter what else we do with the data, _we must preserve rows and columns, or else we lose the data_. If I monkey around and move individual cells haphazardly, I could very well make the spreadsheet apply a predicate falsely. If I do this, I can no longer use the spreadsheet to _learn_ anything about the objects. The spreadsheet will no longer contain _data_ about licenses; because data are, at bottom, just true, reliable predications, we must get the predications right in order to have data. Luckily, pandas will not typically let us modify our data (i.e., our predications).

We _can_, however, slice off various of the predicates we're not interested in. After all, a series of licenses simply named, without any predicates, is still a kind of data: it says that those particular licenses exist. This is a limit case, of course, with one predicate (i.e., existence, unless you think existence isn't a predicate, which is fine), but we are usually interested in cases with at least one additional predicate (and so, at least one additional column). We can, for example, just look at one dimension in the hyperspace: issue date.

```python
df['issuedate'].head(4)
```

We can see that though we sliced off all the dimensions except 'issuedate', we still have the OGCFID index to identify the licenses. Were we to know that the licenses never move vertically in the spreadsheet, we could visually pick them out by their position, but if not, it's better to have them named, because there is no guarantee that the dimension(s) we consider will pick them out uniquely (if, e.g., two licenses were issued on the same date, we cannot use their issue dates as unique identifiers). To hammer home the distinction between _names_ and _predicates_ (here, feature-values), and so the distinction between an index column and a non-index column, notice that the 'values' operation yields 'just the values', meaning the values and not the names (i.e. indices):

```python
df.values[:4]
```

## 2. Data and Our Interests

Now, all that said, there are situations where we don't care about particular objects at all, but only object-types. In such cases, it's okay to use an index column that doesn't index the objects uniquely. Such cases are, in fact, quite common. Anytime we're interested in the relation between objects of type __A__ and objects of type __B__, we can ignore particular objects and focus on predicates. This makes sense of what happens when we make a 'predicate' column the index column: 

```python
dflicensetype = pd.read_csv('abs-licenses-casnd-cleanerxy.csv', index_col='licensetype')
dflicensetype.head(4)
```

The 'index' column is no longer a unique-indexer, but it suits our purposes, because, e.g., we only care about the question whether licenses of type 51 are more commonly issued now than 20 years ago. And if this is all we are interested in, we don't need to know anything about any particular license unless it is of type 51 and was issued either now or 20 years ago. Further, we might as well slice off all predicate-columns except issue date:

```python
dflicensetype['issuedate'].head(20)
```

## 3. Asking Good Questions

### Making Questions Precise

This is not enough, however, to answer our question. Again, that question is whether licenses of type 51 are more commonly issued now than they were 20 years ago. How do we answer this question with the data? We can't say how to answer the question without further specifying what it means. What would it mean to say, e.g., that type-51 licenses _are_ more commonly issued now than 20 years ago? Do we check how many were issued _today_, and how many exactly 20 years ago today, and compare? Well, we have no data about _today_, so let's just treat the most recent date as today.

How do we find the most recent date? Let's order them by value (preserving the index-value link, of course), and look at the tail.

```python
dflicensetype['issuedate'].order().tail()
```

Okay, so the most recent date is July 5, 2013, and on that day, which we are calling 'today', the number of type-51 licenses issued is zero: there were two licenses issued, but neither were of type-51. Well, there cannot possibly be _fewer than zero_ type-51 licenses issued on a given day, so we know that, 20 years ago today, as many or more type-51 licenses were issued as were issued today. How do we find out how many were issued 20 years ago today? If we could see the whole ordered issuedate column, we would simply look for July 5, 1993 and check whether there were any type-51 licenses issued, if any at all. Let's try:

```python
dflicensetype['issuedate'].order()
```

Darn. We can't see the part of the column we need. It's neither at the head nor tail, but somewhere in between, and there are 4987 licenses total. All we need are the rows, if any, with the date 1993-07-05, and their associated indices, which here are license-types. How do we get them? First let's try a method out on rows we can already see, like the last two above:

```python
dflicensetype['issuedate'].loc[dflicensetype['issuedate'] == '2013-07-05T']
```

We got what expected: the two final rows in the spreadsheet, identifying licenses issued today. Ok, now let's swap out the 2013 for 1993:

```python
dflicensetype['issuedate'].loc[dflicensetype['issuedate'] == '1993-07-05T']
```

We got nothing. Now we know that no licenses were issued on July 5, 1993. _A fortiori_, no type-51 licenses were issued on that day.

A lesson about data analysis presents itself here: having a precise question can really sharpen your search through the data. In this way, a precise question functions like attention does in perception: it focuses your analytic resources to a restricted domain to improve performance. Before we precisely specified our question, we didn't really know how to proceed. Afterward, answering the question was trivially easy and very fast.

### Assessing Our Specifications

There is, however, something sort of unsatisfying about our answer above. We get the sneaking feeling that although we answered one very precise form of our question, it wasn't really the same question we asked originally. And unfortunately, many interesting questions about data lack obvious specifications. This circumstance is rife in science, and is usually discussed under the heading 'operationalization'. To operationalize a variable is to give it precise specification, one that can guide experimental intervention. Say, for example, that we are interested in whether or not frequent anxiety is correlated with lower GPA at graduation. The latter variable, GPA at graduation, is extremely precise already (and easy to assess). The former, however, is quite imprecise. What is 'anxiety'? How do we test for it? What counts as 'frequent'? We cannot carry out our research until we find answers to these questions. And even after we answer these questions and conduct our research, it is still open to anyone to critically assess our operationalizations. In the present case, our 'operationalization' is pretty bad. Surely we're not interested in two isolated days, twenty years apart. 

### Informing Action

But what _are_ we interested in here? Say that we found out that in the early 1990s, license type-51 was more popular than it is 'today', where today refers to, say, the past few years. Why would we care to know that? Well, either we are just curious, or knowing that fact somehow informs our purposes and projects. This latter reason picks out what is probably the _primary_ function of data analysis: to inform action (unlike science, whose goal is, nominally at least, simply to find the truth about some domain). And this extends even further the metaphor between data analysis and biological perception. Our particular perceptual abilities vary according to the nature of our projects (at the level of species), which projects are, almost certainly, related to survival and reproduction (this as long as we believe that perceptual systems _adapt_ in the Darwinian sense). Thus, biological perception wouldn't exist were it not for its pragmatic function. And it's helpful, I think, too, to adopt pragmatism as a guiding principle of data analysis. 

Let's consider a somewhat silly but not entirely implausible example: say that you are moving to the San Diego area, and you want to live within the densest cluster of liquor licenses. Say, for example, that you view the density of liquor licenses as a suitable proxy for active nightlife, and you want to live where the action is. Since the present dataset is geocoded, you can use it to find an answer to your (fairly precise) question.

## 4. Where Is the Densest Cluster of Liquor Licenses?

For the time, being, let's ignore the differences in types of licenses. We will revisit differences in license-types later on. All we want to do now is to use the geocode information in the dataset to find the densest cluster of licenses and to be able to name the neighborhood that hosts it. Let's look again at our data:

```python
df.head()
```

First, let's see how many licenses are active.

```python
df.status.value_counts()
```

Let's define a new DataFrame that includes only active licenses. It probably won't change the outcome much, but it's good practice:

```python
dfactive = df.loc[df['status'] == 'ACTIVE']
dfactive.tail()
```

Now let's grab only the columns we need: the geospatial points:

```python
dfactive[['x','y']].head()
```

First, just for fun, let's put these on a plot.

```python
plt.scatter(dfactive.x, dfactive.y,s=1)
plt.xlabel('x')
plt.ylabel('y')
```

This is not helpful. We appear to have one very dense cluster in the upper right, and an outlier or a cluster of outliers in the lower left. What's with the outliers? Let's see how many x values there are in a bin around 0:

```python
dfactive['x'].value_counts(bins=2)
```

Well, there are many 'outliers' (388 to be exact), but not enough to convince me to care about them, since they account for less than 10% of the total points. San Diego does not have this strange bimodal distribution for population, and it's highly unlikely that the distribution of liquor licenses is _so_ radically different from the distribution of population. Thus, there is probably a fluke in the data. [There is another parallel here with biological perception: oftentimes a perceptual system will need to resolve inputs that taken together seem incoherent or unlikely, and it will typically side with the likeliest hypothesis. Here, we eliminate the outliers because given the rest of the information in the dataframe, they are very likely to be erroneous]. And in any case, if I want to live 'near the action', it's happening in the upper right corner. So, let's get rid of the outliers and plot the remaining points.

```python
x = dfactive.x[df.x > 0]
y = dfactive.y[df.y > 1000000]
```

```python
plt.scatter(x, y,s=1)
plt.xlabel('x')
plt.ylabel('y')
```

Aha! Much better, and we can see the San Diego coastline (which means, by the way, that our outliers are out in the Pacific somewhere). But to answer our question with precision, we'll need a better map, one that better distinguishes between higher and lower density areas. Let's also make the plot larger. Here's a map that codes for density using color. The darkest red clusters of points are the densest:

```python
from scipy.stats import gaussian_kde
x = x
y = y
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
fig, ax = plt.subplots(figsize=(20,15))
ax.scatter(x, y, c=z, s=50, edgecolor='')
plt.show()
```

Much better! Visually, the densest clustering is trivially easy to find. Here again is another excellent structural parallel with biological perception. Perception provides immediate access to informationally rich content. Consider how physically complex is the natural scenery in, e.g., a forest. Your visual system processes a massive amount of light information to provide you with a useful representation you can access with ease. The same is true here. Our machine brain, which in this case includes pandas, a kernel density estimator from SciPy, and a visualization engine from matplotlib, has processed nearly 5,000 individual geospatial points and helpfully picked out the densest clusters of these using a color code. Now, we can simply find the darkest red. 

Looks to me like the center of the action is at (6280000, 185000). But let's get a more precise answer, since our data allows it. In the above density algorithm, we computed values for a third dimension, density (z). All we need to do is to find the (x,y) location of the max value for z. This is one of those problems we can solve in many different ways, with many different programming techniques. Here's one that gives us practice using the DataFrame operation in pandas: we create a DataFrame using x, y, and our new density dimension z.

```python
d = {'x': x, 'y': y, 'density': z}
newdf = DataFrame(d)
newdf.head()
```

We know that, being series, x and y retain their indices, and so we don't have to worry about data loss. But how can we be confident that 'z' is indexed in the right way? Let's order the rows of newdf by density values and take a look at the x and y values for the densest points.

```python
newdf.sort(['density']).tail()
```

At a first look, this is promising. The densest points are very close to one another, which we should expect. And my visual estimate of the densest point on the map was (6280000, 1850000). The x and y values at the bottom of this chart are encouragingly close to this estimate. Okay, we're satisfied then. Our density column is indeed a density computation on the geospatial points. And, as it turns out, we now have an answer to the question where is densest cluster of liquor licenses! It's at (6281639.87190, 1849877.89642). Okay, so where is that? We have a format problem. Humans need street addresses, not geospatial coordinates. We need to line up our geospatial coordinates with street addresses. With pandas, this is trivially easy, because our original DataFrame has street addresses. Let's add them to newdf and sort by density as before.

```python
newdf['street address'] = df.premisesaddress
newdf.sort(['density']).tail()
```

And here we have it, folks: of the geospatial points that pick out San Diego establishments with liquor licenses, the one amongst the greatest density of liquor licenses is at 430 Spruce St., which is in Banker's Hill. This makes sense as the center of SD nightlife. 

### Assessing Specifications: Density of Liquor Licenses vs. Density of Bars and Restaurants

Well now hold on a second. The greatest density of liquor licenses in SD is probably a decent measure of nightlife, if for no other reason than it's probably a good measure of population density generally. But it's not the _best_ measure of nightlife available in the data. After all, not all licenses are of the same type. Some are liquor stores, some are bars, some are private clubs, wholesalers, warehouses, etc. Not all of these are relevant for nightlife. If you were looking around the room for socks to wear to the Oscars, you wouldn't settle for any pair of socks. You'll want to look more closely for _dress socks_, because your purposes demand it. Similarly, in this scenario, we need to limit our focus to the license types relevant for active nightlife.

An active nightlife requires an active liquor license, so let's look first at active license types:

```python
dfactive.licensetype.value_counts()
```

The original spreadsheet does not actually describe any of these license types, but the descriptions are available on the CA Alcoholic Beverage Control website. Let's ignore any type with fewer than 86 instances, since they really drop off after that. In San Diego, there are 1591 restaurants that serve beer and wine, 1060 restaurants with full bars, 865 'off-sale' establishments that have beer, wine and booze (like grocery stores and liquor stores), 792 off-sale spots with beer and wine only, 200 bars ('on-sale general public premises'), and 86 beer and wine bars. Bars and resturants conduce to nightlife, and off-sale spots do not, so let's home in on the bars and restaurants. The relevant types are 41, 47, 48, and 42.

Now, we can't just select the rows of newdf with these license types and find the highest value for density, because density in newdf is computed on _all_ license instances. Let's slice off the density column from newdf:

```python
newerdf = newdf[['x','y','street address']]
newerdf.head()
```

And add in license-types:

```python
newerdf['type'] = dfactive.licensetype
newerdf.head()
```

Now we need to get only the rows with types 41, 42, 47 and 48.

```python
nightlife = newerdf[newerdf['type'].isin([41, 42, 47, 48])]
nightlife.head()
```

Okay, _now_ we are ready to run the same analysis we did before.

```python
x2 = nightlife.x
y2 = nightlife.y
x2y2 = np.vstack([x2,y2])
z2 = gaussian_kde(x2y2)(x2y2)
fig, ax = plt.subplots(figsize=(20,15))
ax.scatter(x2, y2, c=z2, s=50, edgecolor='')
plt.show()
```

Not suprisingly, this looks a lot like the last map. But let's find the precise center of the SD nightlife universe.

```python
nightlife['density'] = z2
nightlife.sort(['density']).tail()
```

The epicenter of SD nightlife is...495 Laurel St., which is also in Banker's Hill. Let's find out what place this is.

```python
nightlife['name'] = df.business
nightlife.sort(['density']).tail()
```

## 5. Analysis Without Questions

In the previous section, we considered a question and gave an answer with clear pragmatic import. But as data analysts, we may not always have clear questions to answer. A company or orgranization may simply have a bunch of data and the nebulous goal of 'adding value' to their company. What can we do? Predictably, I've got in mind another parallel with biological perception. Biological perception serves most of the time to inform action, by way of cognition. When we have a well-developed data question, we as data analysts operate like a perceptual system controlled by cognitive attention: we know what we're looking for and this helps us find it. But we are here considering inattentive data 'perception', a kind of analysis made without the guiding hand of a data question. In the biological case, the perceptual system will, with or without attention, continue to provide input to cognition. It will give the same informationally rich, potentially useful picture of the organism's surroundings. 

An organism, like a company or organization, always has at least one unifying end that structures its engagement with the world. Organisms survive and reproduce, companies profit and organizations advance their missions. A biological perceptual system is constituted in such a way that its outputs are the sorts of things that its cognitive system can reliably put in the service of its unifying ends. And even when it operates inattentively, it can provide the organism with information useful for developing more specific projects. For example, a chimp may have the unifying end of obtaining food. It sits on a log, its eyes scanning the nearby environment. Upon seeing a stick of a certain felicitous shape, it forms a plan to use the stick to root out some grubs, which had, until now, been entirely absent from its hierarchy of ends. As data analysts operating without precise data questions, we must behave like inattentive perceptual systems: we must provide 'cognition' with the sorts of inputs it can generally make use of, so that it might act on some unexpected perceptual discovery or start the process of developing a specific plan.

One way to do this is to try to function as much like a perceptual system as possible: to provide informationally-rich and easily accessible data. The best way to do this is by visualization. Let's look at the clustering of licenses over time. We have no special reason to do this, but it may reveal something interesting, or suggest a specific data question.

```python
newerdf['date'] = df.issuedate
newerdf.head()
```

We know how to find nightlife, so let's track the nightlife center over time. This time, let's include a few of the types we ignored last time. Let's make sure, however, to leave out the licenses for bars on boats, since that will shift the nightlife center towards the coast in a perhaps misleading way.

```python
nlbroad = newerdf[newerdf['type'].isin([41, 42, 47, 48, 51, 52, 40, 70, 75])]
```

```python
x3 = nlbroad.x[nlbroad['year'] < '1970']
y3 = nlbroad.y[nlbroad['year'] < '1970']
x3y3 = np.vstack([x3,y3])
z3 = gaussian_kde(x3y3)(x3y3)

x4 = nlbroad.x[nlbroad['year'] < '1980']
y4 = nlbroad.y[nlbroad['year'] < '1980']
x4y4 = np.vstack([x4,y4])
z4 = gaussian_kde(x4y4)(x4y4)

x5 = nlbroad.x[nlbroad['year'] < '1990']
y5 = nlbroad.y[nlbroad['year'] < '1990']
x5y5 = np.vstack([x5,y5])
z5 = gaussian_kde(x5y5)(x5y5)

x6 = nlbroad.x[nlbroad['year'] < '2000']
y6 = nlbroad.y[nlbroad['year'] < '2000']
x6y6 = np.vstack([x6,y6])
z6 = gaussian_kde(x6y6)(x6y6)

x7 = nlbroad.x[nlbroad['year'] < '2010']
y7 = nlbroad.y[nlbroad['year'] < '2010']
x7y7 = np.vstack([x7,y7])
z7 = gaussian_kde(x7y7)(x7y7)

fig, ax = plt.subplots(figsize=(10,7))
ax.scatter(x3, y3, c=z3, s=100, edgecolor='')
plt.title('1969')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.scatter(x4, y4, c=z4, s=100, edgecolor='')
plt.title('1979')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.scatter(x5, y5, c=z5, s=50, edgecolor='')
plt.title('1989')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.scatter(x6, y6, c=z6, s=10, edgecolor='')
plt.title('1999')
plt.show()
```

By 1989, the present pattern was more or less established, but let's look more closely at the early years, because the density center appears to be east of its current point.

```python
nlbroad1969 = {'x': x3, 'y': y3, 'd': z3}
nlb69 = DataFrame(nlbroad1969)
nlb69['address'] = nlbroad['street address']
nlb69['business'] = df.business
nlb69.sort(['d']).tail(1)
```

Ok, now I'm curious. In 1969, the nightlife center was east of the park, in South Park. We know that today, it is west of the park, in Banker's Hill. Let's look at hotspots over time by glancing at the business names at the highest density locations.

```python
nlbroad1969 = {'x': x3, 'y': y3, 'd': z3}
nlb69 = DataFrame(nlbroad1969)
nlb69['address'] = nlbroad['street address']
nlb69['business'] = df.business
nlb69.sort(['d']).tail(20)
```

There's a story to tell here. In 1969, the nightlife (or at least, the drinking center) is distributed all over the region, from Oceanside down to Chula Vista, in private clubs and lodges and veteran's halls. It makes sense that a privatized nightlife will be more distributed: you only go to the club you belong to, so you don't need an entire neighborhood of bars.

```python
nlbroad1979 = {'x': x4, 'y': y4, 'd': z4}
nlb79 = DataFrame(nlbroad1979)
nlb79['address'] = nlbroad['street address']
nlb79['business'] = df.business
nlb79.sort(['d']).tail(20)
```

By 1979, we see a definite shift southward, likely because the drinking culture becomes more public. The Moose Lodge is still the center of the universe, but public restaurants and bars have started to intrude on the top 20. The center is still east of the park despite a growing number of public establishments along the coast. Their locations suggest they are associated with the military.

```python
nlbroad1989 = {'x': x5, 'y': y5, 'd': z5}
nlb89 = DataFrame(nlbroad1989)
nlb89['address'] = nlbroad['street address']
nlb89['business'] = df.business
nlb89.sort(['d']).tail(20)
```

By 1989, the nightlife is starting to coalesce into a tight region west of Balboa Park, roughly in the downtown area. Now, almost all the hotspots are public, and the hottest are Mexican restaurants in Old Town. Long gone are the private clubs and veteran's halls.

```python
nlbroad1999 = {'x': x6, 'y': y6, 'd': z6}
nlb99 = DataFrame(nlbroad1999)
nlb99['address'] = nlbroad['street address']
nlb99['business'] = df.business
nlb99.sort(['d']).tail(20)
```

By 1999, this pattern is firmly established. The midtown/oldtown area near the airport is the place to be. Not a single private venue on the hotlist anymore, and a great deal of ethnic variety has taken hold.

```python
nlbroad2009 = {'x': x7, 'y': y7, 'd': z7}
nlb09 = DataFrame(nlbroad2009)
nlb09['address'] = nlbroad['street address']
nlb09['business'] = df.business
nlb09.sort(['d']).tail(20)
```

The new millenium has seen a pull upwards into the 'hip' neighborhoods of Banker's Hill and Hillcrest. Although the spatial distance between the centers isn't huge, it is notable that _not a single one_ of the midtown/oldtown haunts is on the hotlist. Of course, when the density core is very tight, any spatial shift of that core will repopulate the hotlist very quickly. If the center of nightlife is in Hillcrest, then Hillcrest bars and restaurants are closest to it, and so will turn up on the hotlist. The 'hotlist', as I'm using the term here, is a list of businesses in possession of licenses that are closeby the greatest number of other businesses with licenses. To shift the density center, lots of bars in a new area must obtain licenses, and as they do, each new bar finds itself amongst more close neighbors than did the last. This can ramp up quickly and wipe out the leaderboard.

It is important to keep in mind that we just uncovered a kind of cultural history of San Diego _without ever having any purpose to do so_. We did not begin with a historical question to be answered. One _could_ certainly do that, and social science researchers regularly use data techniques like the ones we are using here. But we came from the other direction. We just plotted a bunch of points on a map, found the densest clusters, and looked at how the center changed over time. Nothing guaranteed from the start our finding anything interesting. It could have been the case that San Diego's nightlife was always downtown, and our 'visual perception' of the density maps would have been a bust. And sometimes it is. Without a data question, your work may sometimes be in vain. But our 'perceptual' strategy - creating rich, immediate visualizations first, before knowing whether it will pay to look closer - is the only reason we looked at the addresses and business types at the density tails over time. It is true that, in order to tell our story in the most interesting way possible, we had to zoom in to the level of the dataframe, but our 'perceptual' information is what cued us to do so. Perception tends to aggregate over a far larger 'dataset' than does cognition. This is advantageous because perception can thereby detect large-scale patterns that deserve a closer look. Of course, when cognition has a specific question to be answered, perception can sometimes serve too as a locus of _detail_, but its function as a _broad_ informant is what helped us here. 

And now that we have a skeleton of a data story to tell, we can use it to a variety of purposes. A hopeful bar owner may be looking for a place to set down roots, and maybe he'd like to set down roots away from the competition, or very near it, or perhaps he'd like to resurrect the Old Town vibe of the 1990s. Maybe a historical organization wants to petition for landmark status of a private lodge. Or perhaps we want only to satisfy our intellectual curiosity about the social landscape of a great American city. The alternatives are legion, and no matter what our final purpose, we are by our methods here poised to answer any of them.
